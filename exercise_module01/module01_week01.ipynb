```python
#Câu 1:
def calc_f1_score (tp,fp,fn):
    precision = tp/(tp+fp)#tính precision
    recall = tp/(tp+fn)#tính recall
    f1_score = (2*precision*recall)/(precision+recall)#tính f1-score
    print(f"precision is {precision}\nrecall is {recall}\nf1-score is {f1_score}")

def result_calc_f1_score(tp,fp,fn): #Đổi tên function từ exercise1 sang result_calc_f1_score
    if not isinstance(tp,int):
        print("tp must be int")
    elif not isinstance(fp,int):
        print("fp must be int")
    elif not isinstance(fn,int):
        print("fn must be int")
    elif tp <= 0 or fp <= 0 or fn <= 0:
        print("tp and fp and fn must be greater than zero")
    else:
        calc_f1_score(tp,fp,fn)

calc_f1_score (tp=2,fp =3,fn =4)
# calc_f1_score (tp='a'fp =3,fn =4)
```

```python
#Câu 3:
import math
import random

# Given
def is_number(num_samples) :
    try :
        int(num_samples) # Type - casting the string to 'float'.
    # If string is not a valid 'float' ,
    # it 'll raise 'ValueError' exception
    except ValueError :
        return False
    return True


def loss_sample_pred_tar(): #Đổi tên function exercise3 sang loss_sample_pred_tar
    num_samples = input("Input number of samples(integer number)which are generated: ")
    if is_number(num_samples) == False:
        print('number of samples must be an integer number')
        return
    num_samples = int(num_samples)
    loss_name = input("Input loss name: ")
    print(f"Input loss name : {loss_name}")

    for i in range(num_samples):
        predict = random.uniform(0, 10)
        target = random.uniform(0, 10)
        if loss_name == 'MAE': #tính MAE
            loss = abs(predict - target)
        if loss_name == 'MSE': #tính MSE
            loss = (predict - target) ** 2
        elif loss_name == 'RMSE': #tính RMSE
            loss = math.sqrt((predict - target) ** 2)
        
        print(f"loss name: {loss_name} , sample: {i} , pred: {predict} , target: {target} , loss: {loss}")
    

# exercise3()
```

```python
#Câu 4:
import math

def approx_cos(x , n):
    result = 0
    for i in range(n):
        coef =(-1) ** i
        num = x **(2 * i) #tử số x^(2n)
        denom = math.factorial(2 * i) #mẫu(2n)!
        term = coef *(num / denom)
        result += term
    return result

def approx_sin(x , n):
    result = 0
    for i in range(n):
        coef =(-1) ** i
        num = x **(2 * i + 1) #tử số x^(2n+1)
        denom = math.factorial(2 * i + 1) #mẫu(2n+1)!
        term = coef *(num / denom)
        result += term
    return result

def approx_sinh(x , n):
    result = 0
    for i in range(n):
        num = x **(2 * i+1) #tử số x^(2n+1)
        denom = math.factorial(2 * i+1) #mẫu(2n+1)!
        term =(num / denom)
        result += term
    return result

def approx_cosh(x , n):
    result = 0
    for i in range(n):
        num = x **(2 * i) #tử số x^(2n)
        denom = math.factorial(2 * i) #mẫu(2n)!
        term =(num / denom)
        result += term
    return result

print(f"approx_sin( x =3.14 , n =10): {approx_sin(x =3.14 , n =10)}")
print(f"approx_cos(x =3.14 , n =10): {approx_cos(x =3.14 , n =10)}")
print(f"approx_sinh(x =3.14 , n =10): {approx_sinh(x =3.14 , n =10)}")
print(f"approx_cosh(x =3.14 , n =10): {approx_cosh(x =3.14 , n =10)}")
```

```python
#Câu 5:
def md_nre_single_sample (y , y_hat , n , p):
    y_root = y ** (1/ n)
    y_hat_root = y_hat ** (1/ n)
    difference = y_root - y_hat_root
    loss = difference ** p
    return loss

print(md_nre_single_sample ( y =100 , y_hat =99.5 , n =2 , p =1))
print(md_nre_single_sample ( y =50 , y_hat =49.5 , n =2 , p =1))
print(md_nre_single_sample ( y =20 , y_hat =19.5 , n =2 , p =1))
print(md_nre_single_sample ( y =0.6 , y_hat =0.1 , n =2 , p =1))
```
